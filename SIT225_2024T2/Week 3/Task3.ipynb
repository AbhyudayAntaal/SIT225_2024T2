```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv(r'C:\Users\abhyu\accelerometer_data.csv')

# Convert the timestamp column to datetime
data['timestamp'] = pd.to_datetime(data['timestamp'])


# Combined plot for all three axes
plt.figure(figsize=(12, 6))
plt.plot(data['timestamp'], data['accel_x'], label='Accel X', color='blue')
plt.plot(data['timestamp'], data['accel_y'], label='Accel Y', color='orange')
plt.plot(data['timestamp'], data['accel_z'], label='Accel Z', color='green')
plt.xlabel('Time')
plt.ylabel('Acceleration')
plt.title('Combined Accelerometer Data')
plt.legend()
plt.show()
```
